```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data1=pd.read_csv("covid19_Confirmed_dataset.csv")
```

```python
data1.head()
```

```python
data1.drop(["Lat","Long"],axis=1,inplace=True)
```

```python
data2=data1.groupby("Country/Region").sum()
```

```python
data2.head()
```

```python
data2.shape
```

```python
plt.figure(figsize=(40,20))
for i in range(187):
    data2.iloc[i].plot()
plt.legend()
plt.xlabel("time line")
plt.ylabel("no. of cases")
plt.show()
```

```python
data2.iloc[0].diff().plot()
```

```python
data2.iloc[0].diff().max()
```

```python
data2.loc["Italy"].diff().max()
```

```python
data2.loc["Spain"].diff().max()
```

```python
countries=list(data2.index)
max_inflection=[]
for i in countries:
    max_inflection.append(data2.loc[i].diff().max())
max_inflection
    
```

```python
data2["max_inflection"]=max_inflection
```

```python
data3=pd.DataFrame(data2["max_inflection"])
data3.head()
```

```python
data4=pd.read_csv("worldwide_happiness_report.csv")
data4.columns
```

```python
useless_cols=["Overall rank","Score","Generosity",'Perceptions of corruption']
```

```python
data4.drop(useless_cols,axis=1,inplace=True)
data4.head()
```

```python
data4.set_index('Country or region',inplace=True)
data4.head()
```

```python
data5=data3.join(data4,how="inner")
data5.head()
```

```python
data5.corr()
```

```python
sns.heatmap(data5.corr(),annot=True)
```
