# centrality analysis: centrality tells the importance of a node in a network,there are multiple ways of determining.

```python
degcent=sorted([i for i in nx.degree_centrality(g).values() ],reverse=True)
deg_cent=nx.degree_centrality(g)
print(degcent)
deg_cent
```

```python
closecent=nx.closeness_centrality(g)
import operator
sorted(closecent.items(),key=operator.itemgetter(1),reverse=True)
```

```python
import operator
btwn_cent=nx.betweenness_centrality(g,normalized=True,endpoints=False)
btwn_cent_approx=nx.betweenness_centrality(g,normalized=True,endpoints=False,k=2)
sorted(btwn_cent.items(),key=operator.itemgetter(1),reverse=True)
```

```python
pg=nx.pagerank(g)
import operator
sorted(pg.items(),key=operator.itemgetter(1),reverse=True)
```

```python
pg_alpha=nx.pagerank(g,alpha=0.85)
import operator
sorted(pg_alpha.items(),key=operator.itemgetter(1),reverse=True)
```
